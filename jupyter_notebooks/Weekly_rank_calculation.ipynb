```python
import pandas as pd
```

```python
pd = pd.read_csv("../resources/billboardHot100_1999-2019.csv")
pd.head(10)
```

```python
df = pd[["Artists", "Name", "Weekly.rank", "Genre"]]
df.head(5)
```

```python
short = df.groupby(['Artists', 'Name', 'Genre']).min()
```

```python
short = short.reset_index()
```

```python
short.to_csv("../resources/weekly_rank.csv", index = False, header=True)
```
