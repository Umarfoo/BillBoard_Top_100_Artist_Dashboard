```python
import pandas as pd
import numpy as np
```

```python
path1 = "Resources/artistDf.csv"
path2 = "Resources/billboardHot100_1999-2019.csv"
path3 = "Resources/songAttributes_1999-2019.csv"
```

```python
artists = pd.read_csv(path1)
billboard = pd.read_csv(path2)
attributes = pd.read_csv(path3)
```

```python
artists_df = artists[['Artist', 'Followers', 'Genres', 'NumAlbums', 'YearFirstAlbum', 'Group.Solo']]
artists_df=artists_df.sort_values(by='Followers', ascending=False)
artists_df=artists_df.reset_index(drop=True)
artists_df.head()
```

```python
artists_DF = artists_df.loc[artists_df['Group.Solo'] == 'Group']
artists_DF = artists_DF.sort_values(by='Followers', ascending = False)
artists_DF.head(6)
```

```python
artists_hmtl = artists_DF.head(6).to_html(index=False)
print(artists_hmtl)
```

```python
artists_hmtl = artists_df.head().to_html(index=False)
print(artists_hmtl)
```

```python
attributes_df = attributes[['Artist','Name','Duration','Popularity', 'Acousticness', 'Danceability', 'Energy', 'Instrumentalness', 'Liveness', 'Loudness', 'Speechiness', 'Tempo', 'Valence']]
attributes_df = attributes_df.round(decimals=2)
attributes_df.sample(n = 5).reset_index(drop=True) 
```

```python
attributes_hmtl = attributes_df.sample(n = 5).reset_index(drop=True).to_html()
print(attributes_hmtl)
```

```python
billboard_df = billboard[['Artists', 'Name', 'Genre', 'Date']]
billboard_df.head()
```

```python
billboard_hmtl = billboard_df.head().to_html()
print(billboard_hmtl)
```
